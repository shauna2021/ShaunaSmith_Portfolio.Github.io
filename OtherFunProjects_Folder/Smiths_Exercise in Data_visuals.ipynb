<u> Histogram in Python </u>

```python
plt.figure(figsize=(11,6))
c=('aqua',
'aquamarine',
'azure')
pickypack=edf[['reading','math','writing']]
_=plt.hist(pickypack, color = c, ec="red")
_=plt.title("Scores For Reading, Math, & Writing In Different States of America", fontsize=15)
_=plt.xlabel('Scores')
_=plt.xticks(rotation=45)
```

<u> Boxplot in Python </u>

```python
plt.figure(figsize=(11,6))
states=crdf.drop(index=0)
badstate=states[['murder','forcible_rape']]
_=sb.boxplot(data=badstate, orient="h", palette="Set1")
_=plt.title("Murder & Rape for States", fontsize=15)
_=plt.xlabel('Counts in States')
_=plt.xticks(rotation=45)
```

<u> Bullet chart in Python </u>

```python
data=edf[['reading','math','writing']]
Adata=edf.math.mean()
Bdata=edf.reading.mean()
Cdata=edf.writing.mean()

Adata, Bdata, Cdata
```

```python
colors = ["blue", "cyan", "yellow", "orange", "red"]

fig = plt.figure(figsize=(11,5))

ax = fig.add_subplot(1,1,1)

plt.barh(y=[0,0,0,0,0], width=[2,2,2,2,2], left=[0,2,4,6,8], color=colors);
plt.barh(y=[0], width=[7.5], color="beige", height=0.2);

plt.vlines(x=9.0, ymin=-0.2, ymax=0.2, color="purple", linewidth=5);

plt.xticks(range(0,11), ["{} %".format(i) for i in range(0,101,10)]);
plt.yticks([0], ["Evaluation"], fontsize=16, fontweight="bold");

ax.spines[["bottom", "top", "left", "right"]].set_visible(False);

plt.xlabel("Percent (%)", fontsize=16, fontweight="bold");
plt.title("Bullet Chart", loc="left", pad=15, fontsize=25, fontweight="bold");
```

```python
%matplotlib inline
limits = [200, 400, 600]
data_to_plot = ("Writing & Math Means: ", 520, 538)
palette = sb.color_palette("Blues_r", len(limits))
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_yticks([1])
ax.set_yticklabels([data_to_plot[0]])

prev_limit = 0
for idx, lim in enumerate(limits):
    ax.barh([1], lim-prev_limit, left=prev_limit, height=35, color=palette[idx])
    prev_limit = lim
    
ax.barh([1], data_to_plot[1], color='purple', height=5)
ax.axvline(data_to_plot[2], color="gray", ymin=0.10, ymax=0.9)
```
