```python
# Begin with Imports of Milestone 1 - 3:

import pandas as pd

Superstoredf=pd.read_csv('C:/Users/Shaun/OneDrive/Superstore_Samples.csv', encoding='unicode_escape')

Superstoredf=Superstoredf.drop(['Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
       'Customer ID', 'Customer Name', 'Country', 'City', 'State',
       'Postal Code', 'Product ID', 'Product Name'], axis=1)

ldf=Superstoredf

SRdrop_index=ldf[ldf['Region']=='South'].index

ldf=ldf.drop(SRdrop_index)
```

```python
correspondence={'West': 3,
               'East': 2,
               'Central': 1}
ldf['Region_code']=ldf.Region.map(correspondence)

correspondence2={'Office Supplies':3,
                'Furniture':2,
                'Technology':1}
ldf['Category_code']=ldf.Category.map(correspondence2)

correspondence3={'Corporate':3, 'Consumer':2, 'Home Office':1}
ldf['Consumer_code']=ldf.Segment.map(correspondence3)

correspondence4={'Labels':17, 'Furnishings':16, 'Art':15, 'Phones':14, 'Binders':13, 'Appliances':12,
       'Tables':11, 'Storage':10, 'Chairs':9, 'Accessories':8, 'Bookcases':7,
       'Envelopes':6, 'Paper':5, 'Fasteners':4, 'Supplies':3, 'Machines':2,
       'Copiers':1}
ldf['Item_id']=ldf['Sub-Category'].map(correspondence4)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_classif

ppdf=ldf.drop(['Segment','Category','Region', 'Sub-Category'], axis=1)
ppdf.head(1)

features=ppdf.loc[:,['Sales','Quantity','Region_code','Category_code','Consumer_code','Item_id']]
target=ppdf.loc[:,'Profit']

fval_sel=SelectPercentile(f_classif, percentile=75)
features_kbest=fval_sel.fit_transform(features,target)

print("Orig num of features: ", features.shape[1])
print("Reduced num of features are: ", features_kbest.shape[1])

ppdff=ppdf.astype(int)

featuresf=ppdff.loc[:,['Sales','Quantity','Region_code','Category_code','Consumer_code','Item_id']]
targetf=ppdff.loc[:,'Profit']

ffval_sel=SelectKBest(f_classif, k=2)
featuresf_kbest=ffval_sel.fit_transform(featuresf,targetf)

print("Orig num of features: ", featuresf.shape[1])
print("Reduced num of features are: ", featuresf_kbest.shape[1])

print("Following features for F-values :\n", features_kbest)
print("Following features by Top 75% :\n", featuresf_kbest)

import numpy as np
def find_outlier(df,x):
    q1=df[x].quantile(.25)
    q3=df[x].quantile(.75)
    iqr=q3-q1
    df['quantiles_score']=np.where(df[[x]]<(q1-1.5*iqr),-1,
                                  np.where(df[[x]]>(q3+1.5*iqr),-1,1))
    return df

find_outlier(ldf, 'Profit')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

_=plt.boxplot(ldf.Profit)

find_outlier(ldf, 'Sales')

_=plt.boxplot(ldf.Sales)

find_outlier(ldf, 'Quantity')

_=plt.boxplot(ldf.Quantity)

_=plt.boxplot(ldf.Category_code)

_=plt.boxplot(ldf.Consumer_code)

#Checking data for duplicates and reasons for
ldf.duplicated().value_counts()

dup_check=ldf[ldf.duplicated()]

DF=ldf.merge(dup_check, indicator=True, how='outer')

defined_indicies=DF[DF['_merge']=='both'].index

DF=DF.drop(defined_indicies)

DF=DF.drop(['_merge'], axis=1)

DF.reset_index(inplace=True)

DF=DF.drop(['index'], axis=1)

find_outlier(DF, 'Profit')
```

```python
#discover and remove outliers 
def outliers(df,ft):
    Q1=df[ft].quantile(0.25)
    Q3=df[ft].quantile(0.75)
    IQR=Q3-Q1
    lower_bounds=Q1-1.5*IQR
    upper_bounds=Q3+1.5*IQR
    ls=df.index[(df[ft]<lower_bounds) | (df[ft]>upper_bounds)]
    return ls

indexed_ls=[]
for feature in ['Sales','Profit']:
    indexed_ls.extend(outliers(DF, feature))

def remove_out(df,ls):
    ls=sorted(set(ls))
    df=df.drop(ls)
    return df

DFC=remove_out(DF,indexed_ls)
DFC.shape
```

```python
# Should we desire to attain Dimensionality Reduction for continued feature extraction or feature selection.
#The following looks at F-values from the target of 'Profit'.

featuresF=DFC.loc[:,['Sales','Quantity','Region_code','Category_code','Consumer_code','Item_id']]
targetF=DFC.loc[:,'Profit']

Fval_sel=SelectKBest(f_classif, k=2)
featuresF_kbest=Fval_sel.fit_transform(featuresF,targetF)

print("Orig num of features: ", featuresF.shape[1])
print("Reduced num of features are: ", featuresF_kbest.shape[1])

DFC.isnull().sum()

discount_dummy=pd.get_dummies(DFC, columns=['Discount'])

discount_dummy.loc[DFC.Discount >= 0.5, 'DiscountPresence_Greater_Half']='Equal or Greater than 50% discount present'

discount_dummy.loc[DFC.Discount < 0.5, 'DiscountPresence_Lesser_Half']='Less than 50% discount present'

discount_dummy.fillna('Not Present').head(2)

marketing_discount_data=discount_dummy.groupby(['Region','Sub-Category']).apply(lambda x: x.groupby('DiscountPresence_Lesser_Half').sum())
marketing_discount_data.sort_values(by = 'Profit', ascending=False).head(2)


HMrefined_df=DFC.drop(['Region_code','Consumer_code'], axis=1)

Highest_market_Regions=HMrefined_df.groupby(['Region','Sub-Category']).apply(lambda x: x.groupby('Profit').sum())
Highest_market_Regions.sort_values(by = 'Sales', ascending=False).head(2)

categorized=DFC.groupby(['Segment']).apply(lambda x: x.groupby('Sub-Category').sum())
categorized.sort_values(by = 'Sub-Category', ascending=False).head(2)

categorized.sort_values(by = 'Sub-Category', ascending=False).tail(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import tree
import warnings
warnings.filterwarnings('ignore')


#first is the feature target for the aspect of 'Consumer_code', 
#which is assigned to 3 values representing the dataset segments of ("Consumer","Corporate","Home Office")
x=NDF.drop(columns='Consumer_code', axis=1)
y=NDF['Consumer_code']

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=10, stratify=y)

#As discrete variables, the selection of Logistic Regression will be applied to the models,
#and 'lbfgs' solver was used due to the dataset size, with max-iter on random_state, set for 10,000 for the solver to converge.
model=LogisticRegression(solver='lbfgs', max_iter=10000, n_jobs=-1)
model.fit(x_train, y_train)

predictions=model.predict(x_test)

#Unfortunately, not much better than guessing, but noted improvement.
score=accuracy_score(y_test,predictions)
score
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.model_selection import GridSearchCV

NDF.describe()

#Attempts thru the Knn of the model by 'Region_code' with hyperparameters to boost performance.
x3=NDF.drop(columns='Region_code', axis=1)
y3=NDF['Region_code']

x3_train,x3_test,y3_train,y3_test=train_test_split(x3,y3, test_size=0.2, random_state=42, stratify=y)

mms=MinMaxScaler()
knn=KNeighborsClassifier(n_neighbors=3)
pipe=Pipeline([('mms', mms),('knn', knn)])

x3_train_scaled=mms.fit_transform(x3_train)
x3_test_scaled=mms.transform(x3_test)

model3=LogisticRegression(solver='lbfgs', max_iter=10000, n_jobs=-1)
model3.fit(x3_train, y3_train)

knn.fit(x3_train_scaled,y3_train)

#assigned a search space by internal identifiers and groups of 3 
search_spaces=[{'knn__n_neighbors':[0,1,2,3,6,9,12,15,18,21]}]
#giving a cross validtaion of 15, with scoring for accuracy
classifierknn = GridSearchCV(pipe, search_spaces, cv=15, verbose=0, 
                             scoring='accuracy').fit(x3_train_scaled, y3_train)

classifierknn.best_estimator_.get_params()["knn__n_neighbors"]

knnGridSearchResults=classifierknn.fit(x3,y3)
knnGridSearchResults.cv_results_['params']

classifierknn.best_score_

classifierknn.best_params_

predicts_knn=knn.predict(x3_test_scaled)
```
