```python
def EvalCdf(t,x):
    count=0.0
    for value in t:
        if value <= x:
            count += 1
        
    prob= count/len(t)
    return prob
```

```python
def PercentileRank(scores, your_score):
    count=0
    for score in scores:
        if score <= your_score:
            count += 1
            
    percentile_rank=100.0*count/len(scores)
    return percentile_rank
```

```python
def Percentile(scores, percentile_rank):
    scores.sort()
    index=percentile_rank*(len(scores)-1)/100
    return scores[index]
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
class HypothesisTest(object):

    def __init__(self, data):
        self.data = data
        self.MakeModel()
        self.actual = self.TestStatistic(data)

    def PValue(self, iters=1000):
        self.test_stats = [self.TestStatistic(self.RunModel()) 
                           for _ in range(iters)]

        count = sum(1 for x in self.test_stats if x >= self.actual)
        return count / iters

    def TestStatistic(self, data):
        raise UnimplementedMethodException()

    def MakeModel(self):
        pass

    def RunModel(self):
        raise UnimplementedMethodException()
```

```python
class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

```python
#Reminder, we are establishing an inference in differences rather than the correlated. 
#Therefore, the less the correlation the greater the established hypothesis in the given probability to the change from fear
class DiffMeansResample(DiffMeansPermute):
   
    def RunModel(self):
      
        group1 = np.random.choice(self.pool, self.n, replace=True)
        group2 = np.random.choice(self.pool, self.m, replace=True)
        return group1, group2
  
```

```python
fear = feardeath['Diseases of Heart'].values #years of deaths during pandemic related to heart disease
others = yearlydeath['Diseases of Heart'].values #years of deaths pre-pandemic related to heart disease
```

```python
def RunResampleTest(fear, others):
   
    data = fear, others
    ht = DiffMeansResample(data)
    p_value = ht.PValue(iters=10000)
    print('\ndiff means resample dofh')
    print('p-value =', p_value)
    print('actual =', ht.actual)
    print('ts max =', ht.MaxTestStat())

    data = (fear,
            others)
    ht = DiffMeansPermute(data)
    p_value = ht.PValue(iters=10000)
    print('\ndiff means resample dofh')
    print('p-value =', p_value)
    print('actual =', ht.actual)
    print('ts max =', ht.MaxTestStat())
```

```python
RunResampleTest(fear, others)
```

Even when we account for the ts max differences between environments.

```python
from thinkstats2 import Mean, MeanVar, Var, Std, Cov

def LeastSquares(xs, ys):
    meanx, varx = MeanVar(xs)
    meany = Mean(ys)

    slope = Cov(xs, ys, meanx, meany) / varx
    inter = meany - slope * meanx

    return inter, slope
```

```python
def FitLine(xs, inter, slope):
    fit_xs = np.sort(xs)
    fit_ys = inter + slope * fit_xs
    return fit_xs, fit_ys
```

```python
def Residuals(xs, ys, inter, slope):
    xs = np.asarray(xs)
    ys = np.asarray(ys)
    res = ys - (inter + slope * xs)
    return res
```

```python
def SampleRows(df, nrows, replace=False):
  
    indices = np.random.choice(df.index, nrows, replace=replace)
    sample = df.loc[indices]
    return sample

def ResampleRows(df):
   
    return SampleRows(df, len(df), replace=True)
```

```python
def SamplingDistributions(reganalyze, iters=101):
    t = []
    for _ in range(iters):
        sample = ResampleRows(ddf)
        overdose = sample.Overdose
        suicides = sample.Suicide
        estimates = LeastSquares(overdose, suicides)
        t.append(estimates)

    inters, slopes = zip(*t)
    return inters, slopes
```

```python
def Summarize(estimates, actual=None):
    mean = Mean(estimates)
    stderr = Std(estimates, mu=actual)
    cdf = thinkstats2.Cdf(estimates)
    ci = cdf.ConfidenceInterval(90)
    print('mean, SE, CI', mean, stderr, ci)
```

```python
iters = 1000
estimates = [ResampleRows(do).mean()
             for _ in range(iters)]
Summarize(estimates)
```

```python
iters = 1000
estimates = [ResampleRows(suicide).mean()
             for _ in range(iters)]
Summarize(estimates)
```

```python
def PlotConfidenceIntervals(xs, inters, slopes, percent=90, **options):
    fys_seq = []
    for inter, slope in zip(inters, slopes):
        fxs, fys = FitLine(xs, inter, slope)
        fys_seq.append(fys)

    p = (100 - percent) / 2
    percents = p, 100 - p
    low, high = thinkstats2.PercentileRows(fys_seq, percents)
    thinkplot.FillBetween(fxs, low, high, **options)
```
